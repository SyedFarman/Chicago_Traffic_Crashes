```python
import pandas as pd
from datetime import datetime

# Read the CSV file using pandas and the detected encoding
df = pd.read_csv('Resources/Traffic_Crashes.csv')

# Access the data as a pandas DataFrame object
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Remove the unwanted columns and any leading/trailing spaces
cols_to_remove = ["CRASH_RECORD_ID","CRASH_DATE_EST_I","LANE_CNT","REPORT_TYPE","INTERSECTION_RELATED_I",
                  "NOT_RIGHT_OF_WAY_I","HIT_AND_RUN_I","DATE_POLICE_NOTIFIED","SEC_CONTRIBUTORY_CAUSE","STREET_NO",
                  "STREET_DIRECTION","STREET_NAME","BEAT_OF_OCCURRENCE","PHOTOS_TAKEN_I","STATEMENTS_TAKEN_I",
                  "DOORING_I","WORK_ZONE_I","WORKERS_PRESENT_I","INJURIES_FATAL","INJURIES_INCAPACITATING",
                  "INJURIES_NON_INCAPACITATING","INJURIES_REPORTED_NOT_EVIDENT","INJURIES_NO_INDICATION","INJURIES_UNKNOWN",
                 "LOCATION"]
df = df.drop([col.strip() for col in cols_to_remove if col.strip() in df.columns], axis=1)

# Print the resulting DataFrame
df.head()
```

```python
df.info()
```

```python
# Replace NaN values with 0 & NA
df['WORK_ZONE_TYPE'] = df['WORK_ZONE_TYPE'].fillna(value='NA')
df['NUM_UNITS'] = df['NUM_UNITS'].fillna(0)
df['MOST_SEVERE_INJURY'] = df['MOST_SEVERE_INJURY'].fillna(value='NA')
df['INJURIES_TOTAL'] = df['INJURIES_TOTAL'].fillna(0)
df['LATITUDE'] = df['LATITUDE'].fillna(0)
df['LONGITUDE'] = df['LONGITUDE'].fillna(0)
# Print the resulting DataFrame
df.head()
```

```python
df.info()
```

```python
# Rename the columns
new_column_names = {'FIRST_CRASH_TYPE': 'CRASH_REASON','ALIGNMENT':'ROAD_ALIGNMENT','PRIM_CONTRIBUTORY_CAUSE': 'CONTRIBUTORY_CAUSE', 
                    'MOST_SEVERE_INJURY': 'INJURY_TYPE'}
df = df.rename(columns=new_column_names)

# Print the resulting DataFrame
df.head(5)
```

```python
# get the sorted CRASH_DATE
Sort_Dates = df['CRASH_DATE'].sort_values()

# display the result
print(Sort_Dates)
```

```python
# convert CRASH_DATE column to datetime datatype - UTC
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], format='%m/%d/%Y %I:%M:%S %p')

# extract year from CRASH_DATE column and create new CRASH_YEAR column
df['CRASH_YEAR'] = df['CRASH_DATE'].dt.year
```

```python
# Replacing values of Injury_Cat Column with 0 & 1

df['INJURY_CAT'] = df['CRASH_TYPE']

df['INJURY_CAT'] = df['INJURY_CAT'].replace(['NO INJURY / DRIVE AWAY', 'INJURY AND / OR TOW DUE TO CRASH'], [0, 1])
df.head(5)
```
